```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
data = {'Categoría': ['A', 'B', 'C', 'D'],
        'Valores': [30, 45, 15, 10]}

df = pd.DataFrame(data)

# Crear un gráfico de pastel
plt.figure(figsize=(16, 8))
plt.pie(df['Valores'], labels=df['Categoría'], autopct='%1.1f%%', startangle=140)
plt.title('Gráfico de Pastel Ejemplo')
plt.axis('equal')  # Esto hace que el gráfico sea circular
plt.show()
```
